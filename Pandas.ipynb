**Group By**

```python
data = {
    'Name': ['Sam','Abhe','Davis','John'],
    'Company': ['Google', 'FB','FB','IBM'],
    'Sales': [100,567,389,123],
}
```

```python
df = pd.DataFrame(data)
df
```

```python
bycomp = df.groupby('Company')
bycomp.mean()
```

```python
bycomp.sum()
```

```python
df
```

```python
bycomp.std()
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').describe()
```

Merging 

```python
a = {
    'key':['s1','s1','s2','s4'],
    'A':[10,20,30,40],
    'B':[1,2,3,4]
}
```

```python
b = {
    'key':['s1','s2','s3','s4'],
    'C':[100,200,300,400],
    'D':[11,12,13,14]
}
```

```python
a = pd.DataFrame(a)
a
```

```python
b = pd.DataFrame(b)
b
```

```python
pd.merge(a,b, on='key')
```

Joining

```python
a = {
    'A':['A1','A2','A3','A4','A5'],
     'B':['B1','B2','B3','B4','B5']
} 
```

```python
b = {
    'C':['C1','C2','C3','C4','C5'],
     'D':['D1','D2','D3','D4','D5']
}
```

```python
a = pd.DataFrame(a, index= ['M0','M1','M2','M3','M4'])
a
```

```python
b = pd.DataFrame(b, index= ['M0','M1','M2','M3','M5'])
b
```

```python
b.join(a, how ='inner')
```
