```python
# importing packages 
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()
```

```python
# loading dataset 
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

**Numerical Data plot**

Line plot

```python
sns.lineplot(x="total_bill", y="tip", data=tips);
```

Distribution plot

```python
sns.set(style = 'darkgrid', rc = {'figure.figsize':(10, 6)}, font_scale = 1.5)
```

```python
sns.distplot(tips['total_bill']);
```

```python
sns.distplot(tips['total_bill'], bins = 20)
```

```python
sns.distplot(tips['total_bill'], kde = False);
plt.xlabel('Total bills');
```

```python
sns.distplot(tips['total_bill'], hist = False, vertical= True, color = 'red')
```
